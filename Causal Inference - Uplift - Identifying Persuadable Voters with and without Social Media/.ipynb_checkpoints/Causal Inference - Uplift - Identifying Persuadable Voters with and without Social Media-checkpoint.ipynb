### Creating Stratified Marketing Data Sample to Match Twitter Dataset Size

```python
# Combines treatment and conversion values for single column stratification
marketing_df['stratum'] = (marketing_df['treatment'].astype(str) + "-" + \
                           marketing_df['conversion'].astype(str))
```

```python
# Confirms column creation was successful
marketing_df.head()
```

```python
# Defines samples size to match Twitter dataset
total_n = 26353

def stratified_sample(df, total_n, random_state=0):
    """Uses stratum column to create stratified sample to match Twitter dataset 
    with the same treatment and conversion proportions"""

    # Tallies treatment conversion combination counts
    counts = df['stratum'].value_counts()
    # Calculates proportions for each combination
    props = counts / counts.sum()
    # Calculates desired number of rows per stratum 
    n_per = (props * total_n).round(0).astype(int)  
    # Adjusts first stratum to account for rounding errors 
    n_per.iloc[0] += total_n - n_per.sum()

    samples = [] 
    for s, n in n_per.items(): 
        # Extracts rows belonging to each stratum combination
        group = df[df['stratum'] == s]          
        
        if len(group) >= n: 
            # Samples n rows without replacement 
            sampled = group.sample(n=n, random_state=random_state, replace=False)
        else: 
            # Takes all available rows if stratum is too small 
            sampled = group 
        
        samples.append(sampled)

    # Concatenates all stratum samples into a single df
    result = pd.concat(samples, ignore_index=True)
    return result

# Applies stratified sampling function to marketing data to match Twitter size
marketing_sample = stratified_sample(df=marketing_df, total_n=total_n, 
                                     random_state=0)
```

### Feature Selection for Twitter Data

### Feature Extraction Using BERT 
I ended up using SentenceTransformer, which is a miniLM distilled from BERT. SentenceTransformer is better at sentence embeddings out of the box, as it has been optimized for exactly this, while still sharing the same basic architecture as BERT. 

```python
# Loads BERT-distilled pre-trained model for sentence embeddings
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
# Pulls text data from df to simple list
text_list = text_df["Text"].tolist()

# Extracts embeddings from created list
embeddings = model.encode(text_list, show_progress_bar=True)
```

```python
# Outputs shape for sanity check
print("BERT embeddings shape:", embeddings.shape)
```

```python
# Sets adjustable component amount to a variable
n_components = 100

# Initiates PCA model
pca = PCA(n_components=n_components)

# Uses embeddings to create principal components to maximize variance for uplift
# model
twitter_pca = pca.fit_transform(embeddings)

# Outputs values for sanity check
print("BERT embeddings shape:", embeddings.shape)
print("After PCA:", twitter_pca.shape)
```

```python
# Converts to data frame with relational column names 
twitter_pca_df = pd.DataFrame(twitter_pca, columns=[f'twitter_pca_{i}' for i in 
                                                      range(1, n_components + 1)])
```

### Merge Datasets

```python
# Merges marketing and twitter data 
marketing_and_twitter = pd.concat([marketing_sample.reset_index(drop=True), 
                                   twitter_pca_df.reset_index(drop=True)],
                                  axis=1)
```

```python
# Outputs shape for sanity check after merging 
print("Combined shape:", marketing_and_twitter.shape)
```

```python
# Checks to make sure the new dataframe looks good 
marketing_and_twitter.head()
```

### Train-Test-Split for both Models 
I made sure to again stratify the split for both models, to preserve the treatment and conversion ratios present in the original dataset. Doing the train-test-split is a little different for an uplift model, because of the treatment inclusion necessary to properly evaluate the model. 

```python
# Isolates both target and treatment variables for the first model 
X1 = marketing_sample.drop(columns=['treatment', 'conversion', 'stratum'], axis=1)
y1 = marketing_sample['conversion']
treatment1 = marketing_sample['treatment']
```

```python
# Stratify splits for first model
X1_train, X1_test, y1_train, y1_test, treatment1_train, treatment1_test = train_test_split(
    X1, y1, treatment1, test_size=0.2, random_state=0, stratify=marketing_sample['stratum'])
```

```python
# Isolates both target and treatment variables for the second model 
X2 = marketing_and_twitter.drop(columns=['treatment', 'conversion', 'stratum'], 
                                axis=1)
y2 = marketing_and_twitter['conversion']
treatment2 = marketing_and_twitter['treatment']
```

```python
# Stratify splits for second model
X2_train, X2_test, y2_train, y2_test, treatment2_train, treatment2_test = train_test_split(
    X2, y2, treatment2, test_size=0.2, random_state=0, stratify=marketing_and_twitter['stratum'])
```

### Model & Hyperparameter Selection
First, I defined the models I wanted to choose from. Uplift models are interesting because you're essentially using two models in one. I decided to build my uplift models on Logistic Regression models, for their simplicity and interpretability. I'm evaluating the model on its uplift at 10% score, to optimize uplift for the top 10% of persuadables. 

```python
# Defines uplift models to choose from
models = {
    'Solo Model': SoloModel(LogisticRegression(random_state=0)),
    'Class Transformation': ClassTransformation(LogisticRegression(random_state=0)),
    'Two Models': TwoModels(estimator_trmnt=LogisticRegression(random_state=0),
                            estimator_ctrl=LogisticRegression(random_state=0))
}

cv = KFold(n_splits=5, shuffle=True, random_state=0)
best_models = {}
results = []
```

```python
for dataset_name, (X_train, y_train, treatment_train, X_test, y_test, treatment_test) in {
    'Marketing Only': (X1_train, y1_train, treatment1_train, X1_test, y1_test, treatment1_test),
    'Marketing + Twitter': (X2_train, y2_train, treatment2_train, X2_test, y2_test, treatment2_test)
}.items():
    # Creates dataset-specific scorer inside the loop
    scoring = make_uplift_scorer('uplift_at_k', treatment=treatment_train, k=0.1, 
                                 strategy='overall')
    
    for model_name, model in models.items():
        # Grid search params for each model 
        param_grid = {
            'estimator__C': [0.1, 0.5, 1.0, 2.0, 5.0, 9.36, 10.0],
            'estimator__penalty': ['l2'],
            'estimator__solver': ['liblinear']
        }
        
        if 'Two Models' in model_name:
            param_grid = {
                'estimator_trmnt__C': [0.1, 0.5, 1.0, 2.0, 5.0, 9.36, 10.0],
                'estimator_trmnt__penalty': ['l2'],
                'estimator_trmnt__solver': ['liblinear'],
                'estimator_ctrl__C': [0.1, 0.5, 1.0, 2.0, 5.0, 9.36, 10.0],
                'estimator_ctrl__penalty': ['l2'],
                'estimator_ctrl__solver': ['liblinear']
            }
        
        grid_search = GridSearchCV(
            model, param_grid, 
            cv=cv, scoring=scoring, 
            n_jobs=1, verbose=1, return_train_score=True
        )
        
        # Passes treatment into fit
        grid_search.fit(X_train, y_train, treatment=treatment_train)
        
        best_models[f'{dataset_name} - {model_name}'] = grid_search.best_estimator_
        
        # Manually scores holdout set using best estimator
        holdout_uplift = grid_search.best_estimator_.predict(X_test)
        holdout_score = uplift_at_k(y_test, holdout_uplift, treatment_test, 
                                    strategy='overall', k=0.1)
        
        results.append({
            'Dataset': dataset_name,
            'Model': model_name,
            'Best_Params': grid_search.best_params_,
            'CV Average': f"{grid_search.best_score_:.4f}",
            'Holdout Score': f"{holdout_score:.4f}"
        })
        print(f"{dataset_name} | {model_name}: CV Average = {grid_search.best_score_:.4f}, "
              f"Holdout Score = {holdout_score:.4f}, "
              f"Best Params = {grid_search.best_params_}")
```

### Model Evaluation
The evaluation metrics clearly showed that adding the Twitter data to the marketing campaign data enriched the signal. The evaluation metrics were sorted by their uplift score at 10%, as determined by the selection metric. The marketing-only models had the three lowest scores across all metrics. The holdout scores and CV average scores are close across all models, suggesting minimal overfitting. The training and CV scores for the best model also have a minimal gap, further supporting this idea. The Class Transformation model with the combined datasets had the highest uplift score (0.1030) at the 10% evaluation metric. This can be interpreted to mean that the top 10% of persuadables identified by the model had a 10.30 percentage-point higher conversion rate due to treatment. Although the combined dataset Two Model has the highest Qini AUC score, meaning it's better at ranking persuadables across the entire population, the combined dataset Class Transformation still dominated the Uplift AUC score. All three models for the combined datasets performed better than the random baseline in the first 10% on the Qini curves, and then tapered off in the remaining 90%. The marketing-only models either barely exceeded the random baseline or performed worse than it. Because the purpose of this project was to identify the top 10% of persuadables, the combined Class Transformation model will be used to predict them, given its higher uplift score at 10%, even though it had the weakest Qini curve performance among the three combined models. The class transformation model also scored highest in the uplift AUC metric, with the highest area beneath the uplift curve. The holdout scores and CV average scores are close across all models, suggesting minimal overfitting. The training and CV scores for the best model also have a minimal gap, further supporting this idea 

```python
# Stores results from top model 
best_idx = grid_search.best_index_
train_score = grid_search.cv_results_['mean_train_score'][best_idx]
cv_score = grid_search.best_score_

# Checks for overfitting
print(f"Train: {train_score:.4f} | CV: {cv_score:.4f} | Gap: {train_score-cv_score:.4f}")
```

```python
# Defines datasets from which to test
test_sets = {
    'Marketing Only': (X1_test, y1_test, treatment1_test),
    'Marketing + Twitter': (X2_test, y2_test, treatment2_test)
}

# Creates placeholder for metrics
metrics = []
for name, model in best_models.items():
    dataset_name = name.split(' - ')[0]
    X_test_ds, y_test_ds, treatment_test_ds = test_sets[dataset_name]
    
    uplift_pred = model.predict(X_test_ds)
    
    metrics.append({
        'Model': name,
        'Qini AUC': qini_auc_score(y_test_ds, uplift_pred, treatment_test_ds),
        'Uplift AUC': uplift_auc_score(y_test_ds, uplift_pred, treatment_test_ds),
        'Uplift at 10%': uplift_at_k(y_test_ds, uplift_pred, treatment_test_ds, 
                                     strategy='overall', k=0.1)
    })

evaluation_results = pd.DataFrame(metrics).round(4)
print(evaluation_results.sort_values('Uplift at 10%', ascending=False))
```

```python
# Sets serif font globally for desired design aesthetic
plt.rcParams['font.family'] = ['serif']    
plt.rcParams['font.weight'] = 'light'      

models_to_plot = [
    'Marketing + Twitter - Class Transformation',
    'Marketing + Twitter - Two Models',
    'Marketing + Twitter - Solo Model',
    'Marketing Only - Two Models',
    'Marketing Only - Solo Model',
    'Marketing Only - Class Transformation'
]

# Creates shortened labels to preserve space on plots
model_labels = [
    'CT -\nCombined',
    'TM -\nCombined',
    'SM -\nCombined',
    'TM -\nMarketing',
    'SM -\nMarketing',
    'CT -\nMarketing'
]

metrics_to_plot = ["Uplift at 10%", "Qini AUC", "Uplift AUC"]

metric_titles = {
    "Uplift at 10%": "Uplift at 10%",
    "Qini AUC": "Qini AUC",
    "Uplift AUC": "Uplift AUC"
}

# Extracts metric values for each model
vals = {m: [] for m in metrics_to_plot}
for model in models_to_plot:
    row = evaluation_results[evaluation_results["Model"] == model].iloc[0]
    for m in metrics_to_plot:
        vals[m].append(row[m])

# Defines bar chart positioning
x = np.arange(len(model_labels))
width = 0.5

# Defines colors to be used
bar_color = '#4A90B8'
threshold_color = '#E07B39'

# Creates 1x3 subplot grid for the three evaluation metrics
fig, axes = plt.subplots(1, 3, figsize=(18, 7))

legend_elements = [
    Line2D([0], [0], color=bar_color, lw=8, label="Model Score"),
    Line2D([0], [0], color=threshold_color, lw=2, linestyle='dashed', 
           label="Random Baseline (0)")
]

# Plots each metric in its subplot
for i, m in enumerate(metrics_to_plot):
    ax = axes[i]

    bar_colors = [bar_color if 'Combined' in model_labels[j] 
                  else '#A8C8E0' for j in range(len(model_labels))]

    ax.bar(x, vals[m], width, color=bar_colors, alpha=0.85, 
           edgecolor='black', linewidth=0.5)

    # Defines random baseline 
    ax.axhline(0, color=threshold_color, linestyle="dashed", linewidth=2)
    ax.text(len(model_labels) - 0.5, 0, "Random Baseline", 
            ha="right", va="bottom", fontsize=9, fontweight='light', 
            style='italic', color=threshold_color,
            bbox=dict(boxstyle="round,pad=0.3", facecolor='white',
                     alpha=0.9, edgecolor=threshold_color, linewidth=0.5))

    ax.set_xticks(x)
    ax.set_xticklabels(model_labels, rotation=0, ha="center", fontsize=9)
    ax.set_title(metric_titles[m], color='black', fontsize=14, pad=20)
    ax.grid(axis='y', alpha=0.3, linestyle='--')

# Shared legend on first subplot only, to reduce clutter
axes[0].legend(handles=legend_elements, loc="upper right",
               frameon=True, fancybox=True, shadow=True)

plt.suptitle('Uplift Model Evaluation Metrics', fontsize=16, fontweight='light', y=1.02)
plt.tight_layout()
plt.savefig('uplift_evaluation_metrics.png', 
            dpi=300, bbox_inches='tight', facecolor='white')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

dataset_groups = {
    'Marketing + Twitter': ['Marketing + Twitter - Class Transformation',
                            'Marketing + Twitter - Two Models',
                            'Marketing + Twitter - Solo Model'],
    'Marketing Only': ['Marketing Only - Class Transformation',
                       'Marketing Only - Two Models',
                       'Marketing Only - Solo Model']
}

test_sets = {
    'Marketing + Twitter': (X2_test, y2_test, treatment2_test),
    'Marketing Only': (X1_test, y1_test, treatment1_test)
}

# Keeping with similar color scheme 
colors = ['#1A5276', '#4A90B8', '#A8C8E0']
model_short = ['Class Transformation', 'Two Models', 'Solo Model']

plt.rcParams['font.family'] = ['serif']
plt.rcParams['font.weight'] = 'light'

# Creates qini curves for each model from each dataset
for i, (dataset_name, model_keys) in enumerate(dataset_groups.items()):
    ax = axes[i]
    X_test_ds, y_test_ds, treatment_test_ds = test_sets[dataset_name]

    for j, model_key in enumerate(model_keys):
        model = best_models[model_key]
        uplift_pred = model.predict(X_test_ds)
        x_vals, y_vals = qini_curve(y_test_ds, uplift_pred, treatment_test_ds)
        auc = qini_auc_score(y_test_ds, uplift_pred, treatment_test_ds)
        x_vals_norm = x_vals / x_vals.max()
        ax.plot(x_vals_norm, y_vals, color=colors[j], linewidth=2,
                label=f'{model_short[j]} (AUC={auc:.4f})')

    # Orange dashed random baseline and solid orange 10% threshold
    ax.axhline(0, color='#E07B39', linewidth=1.5, linestyle='dashed', label='Random Baseline')
    ax.axvline(0.10, color='#E07B39', linewidth=1.5, linestyle='solid', label='10% Threshold')

    ax.set_title(dataset_name, fontsize=14, pad=15)
    ax.set_xlabel('Fraction of Population Targeted')
    ax.set_ylabel('Qini')
    ax.set_xlim(0, 1)
    ax.legend(loc='lower left', bbox_to_anchor=(0.1, 0.05),
          frameon=True, fancybox=True, shadow=True)
    ax.grid(alpha=0.3, linestyle='--')

plt.suptitle('Qini Curves by Dataset', fontsize=16, fontweight='light', y=1.02)
plt.tight_layout()
plt.savefig('qini_curves.png', dpi=300, bbox_inches='tight', facecolor='white')
plt.show()
```

# Generating Persuadable Rankings
After generating the persuadable rankings, the summary of the top 10% of raw predicted uplift scores shows strong uplift potential (mean: 0.818) compared to random selection. With a minimum of 0.396268, targeting could be expanded slightly based on resource capacity (see Appendix C). You can see that the uplift scores remain well above the random baseline through ~10%, approaching it by 13%.

```python
# Pick the actual best model (highest Uplift@10%)
best_model = best_models[evaluation_results.loc[evaluation_results['Uplift at 10%'].idxmax(), 'Model']]
print(f"Best model loaded: {best_model}")

# Uses best model to predict uplift on full combined dataset
X2_full = marketing_and_twitter.drop(columns=['treatment', 'conversion', 'stratum'])

uplift_scores = best_model.predict(X2_full)

# Adds uplift scores to the full dataset
marketing_and_twitter['uplift_score'] = uplift_scores

# Ranks by uplift score and selects top 10%
marketing_and_twitter_sorted = marketing_and_twitter.sort_values('uplift_score', ascending=False)
top_10_percent = marketing_and_twitter_sorted.head(int(len(marketing_and_twitter_sorted) * 0.10))

print(f"Total population: {len(marketing_and_twitter)}")
print(f"Top 10% persuadables identified: {len(top_10_percent)}")
print(f"\nUplift score summary for top 10%:")
print(top_10_percent['uplift_score'].describe())
```

```python
plt.rcParams['font.family'] = ['serif']
plt.rcParams['font.weight'] = 'light'

fig, ax = plt.subplots(figsize=(14, 6))

# Resets index for clean ranking
top_10_percent_reset = top_10_percent.reset_index(drop=True)
top_10_percent_reset['rank'] = top_10_percent_reset.index + 1

top_1_cutoff = int(len(top_10_percent_reset) * 0.10)

# Colors points by tier — top 1%, 1-5%, 5-10%
tier_colors = ['#1A5276' if r <= top_1_cutoff
               else '#4A90B8' if r <= len(top_10_percent_reset) * 0.50
               else '#A8C8E0'
               for r in top_10_percent_reset['rank']]

ax.scatter(top_10_percent_reset['rank'], top_10_percent_reset['uplift_score'],
           c=tier_colors, alpha=0.6, s=10)

# Random baseline at 0
ax.axhline(0, color='#E07B39', linewidth=1.5,
           linestyle='dashed', label='Random Baseline')

ax.set_xlabel('Rank', fontsize=12)
ax.set_ylabel('Uplift Score', fontsize=12)
ax.set_title('Uplift Score Distribution — Top 10% Persuadables', fontsize=14, pad=20)
ax.grid(alpha=0.3, linestyle='--')

legend_elements = [
    Line2D([0], [0], marker='o', color='w', markerfacecolor='#1A5276',
           markersize=8, label=f'Top 1% of Persuadables (n={top_1_cutoff})'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='#4A90B8',
           markersize=8, label='1-5% of Persuadables'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='#A8C8E0',
           markersize=8, label='5-10% of Persuadables'),
    Line2D([0], [0], color='#E07B39', linewidth=1.5,
           linestyle='dashed', label='Random Baseline')
]

ax.legend(handles=legend_elements, loc='upper right',
          frameon=True, fancybox=True, shadow=True)

plt.tight_layout()
plt.savefig('persuadable_rankings.png', dpi=300, bbox_inches='tight', facecolor='white')
plt.show()
```

## Conclusions and Recommendations
The project demonstrated that using the embedded and compressed 2024 election Twitter data enriched the marketing campaign data sample in a meaningful way. Uplift models are unique in that you often expect minimal signal in noisy data, yet they can still deliver incremental results that create data-driven change when used wisely. The results from combining the marketing campaign and Twitter data suggested that expanding reach to nearly the top 13% of the population is possible, provided that the uplift score × average conversion value is greater than the campaign cost per person.

```python
# Defines top 13% to see where scatter plot lands
top_13_percent = marketing_and_twitter_sorted.head(int(len(marketing_and_twitter_sorted) * 0.13))

plt.rcParams['font.family'] = ['serif']
plt.rcParams['font.weight'] = 'light'

fig, ax = plt.subplots(figsize=(14, 6))

# Resets index for clean ranking — now for top 13%
top_13_percent_reset = top_13_percent.reset_index(drop=True)
top_13_percent_reset['rank'] = top_13_percent_reset.index + 1

top_1_cutoff = int(len(top_13_percent_reset) * 0.10)  
top_5_cutoff = int(len(top_13_percent_reset) * 0.385)  

# Colors points by tier — top 1%, 1-5%, 5-13%
tier_colors = ['#1A5276' if r <= top_1_cutoff
               else '#4A90B8' if r <= top_5_cutoff
               else '#A8C8E0'
               for r in top_13_percent_reset['rank']]

ax.scatter(top_13_percent_reset['rank'], top_13_percent_reset['uplift_score'],
           c=tier_colors, alpha=0.6, s=10)

# Random baseline at 0
ax.axhline(0, color='#E07B39', linewidth=1.5,
           linestyle='dashed', label='Random Baseline')

ax.set_xlabel('Rank', fontsize=12)
ax.set_ylabel('Uplift Score', fontsize=12)
ax.set_title('Uplift Score Distribution — Top 13% Persuadables', fontsize=14, pad=20)
ax.grid(alpha=0.3, linestyle='--')

legend_elements = [
    Line2D([0], [0], marker='o', color='w', markerfacecolor='#1A5276',
           markersize=8, label=f'Top 1% of Persuadables (n={top_1_cutoff})'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='#4A90B8',
           markersize=8, label=f'1-5% of Persuadables (n={top_5_cutoff - top_1_cutoff})'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='#A8C8E0',
           markersize=8, label='5-13% of Persuadables'),
    Line2D([0], [0], color='#E07B39', linewidth=1.5,
           linestyle='dashed', label='Random Baseline')
]

ax.legend(handles=legend_elements, loc='upper right',
          frameon=True, fancybox=True, shadow=True)

plt.tight_layout()
plt.savefig('persuadable_rankings_13pct.png', dpi=300, bbox_inches='tight', facecolor='white')
plt.show()
```
