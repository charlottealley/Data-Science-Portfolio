## Narrative
My original idea was to create a model that predicts whether an individual is living with their parents based on various features. This idea was sparked by my 33-year-old brother’s recent third move home into my mother’s basement. Interestingly, the same article goes on to point out that the cost of living and unemployment did not have a strong link to the rates of young adults living at home. I find this phenomenon to be fascinating and worth further exploration.  My idea ended up evolving into predicting whether the head of household is part of an “extended” or “single” family household when I found the Household Data Set on Kaggle. I chose this set because of the clear target variable and the number of features it contains. 
I believe this problem can be meaningful to urban planners, social researchers, and government agencies that design housing policies and allocate resources based on household composition. Understanding the distribution and features of different household types can help to optimize social services, infrastructure planning, and community development. Currently, many urban planning efforts depend on outdated census numbers or general population statistics without detailed real-time insight into household structures, which can limit precision in decision-making. This project addresses that gap by using machine learning to more reliably classify household types from structured data, enabling for more targeted interventions. 
The dataset chosen contains numerous attributes related to household features and a clear target variable identifying whether each head of household is part of an "extended" or "single family." Although the dataset lacks comprehensive source documentation, it presents an opportunity to simulate data quality challenges that are common in real-world scenarios and apply exploratory data analysis techniques. This exploratory analysis will provide insights into key predictors that separate household types, getting us one step closer to choosing the appropriate model.

```python
import pandas as pd
import numpy as np

# Creates dataframe from file, outputs first five rows
household_df = pd.read_csv('family_info.csv')
household_df.head()
```

```python
# Checks to confirm target variable has only two classes
household_df['Type of Household'].value_counts()
```

```python
# Creates new dataframe with all data except the third class that excludes 
# family members
household_df = household_df[household_df['Type of Household'] 
    != 'Two or More Nonrelated Persons/Members']
household_df['Type of Household'].value_counts()
```

## Graph 1: Distribution of Household Types
I decided to remove all rows containing the third category for the Type of Household target variable (Two or More Nonrelated Persons/Members) before plotting, since this was a very small percentage of the total data (0.40%), and it didn't align with the binary classification model vision for the project. The first graph shows the distribution of household types. It is clear that the classes are imbalanced, with single-family households at almost twice the size of extended-family households. The imbalance will need to be accounted for when creating the model. 

```python
import matplotlib.pyplot as plt

# Plots distribution of target variable classes 
household_df['Type of Household'].value_counts().plot(kind='bar', color='grey')
plt.title('Distribution of Household Types')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

## Graph 2: Household Type by Head of Household Age 
This boxplot tells us that the median age for the head of household for the single-family class is quite a bit lower than the median age for the extended-family class. This aligns with the idea that the expectation that the extended-family head of households are more likely to be of the older generation (meaning adult children are living with their parents, rather than the other way around). There's also a large number of outliers for both categories, which will need to be addressed in the data preparation portion of the assignment. 

```python
# Boxplots for distribution of household type by head of household ages
household_df.boxplot(column='Household Head Age', by='Type of Household', 
                     grid=False, color='grey')
plt.title('Household Type by Head of Household Age')
plt.suptitle('')
plt.xlabel('Household Type')
plt.ylabel('Head Age')
plt.show()
```

## Graph 3: Average Family Size by Household Type
We can see that the average family size for the extended-family class is just below 6 compared to 4 for the single-family class. The higher family size for the extended-family class aligns with expectations, signifying multiple generations living together. The difference between the two is meaningful, especially when imagining how much more cramped a house or rental can get when you add two more people to it.

```python
# Uses groupby to find family size mean, plots average size per class
avg_family_size = household_df.groupby(
    'Type of Household')['Total Number of Family members'].mean()
avg_family_size.plot(kind='bar', color=['grey'])
plt.title('Average Family Size by Household Type')
plt.xlabel('Household Type')
plt.ylabel('Average Family Size')
plt.xticks(rotation=0)
plt.show()
```

## Data Cleaning
It looks like we have two categorical variables that have missing values (about 18%). Since these variables are categorical, I cannot use interpolation. Since the XGBoost model natively handles missing values, I'm choosing to leave them as is. There aren't any duplicate rows or duplicate index values in this set. I standardized the text within the categorical variables to reduce potential text error issues.

```python
# Checks for missing values
household_df.info()
```

```python
# Checks for row duplicates, none exist
household_df.duplicated().sum()
```

```python
# Checks for index duplicates, none exist
household_df.index.duplicated().sum()
```

```python
import re

# Loops through object columns with standardization rules, using method chaining  
# and regular expressions to alter text
for col in household_df.select_dtypes(include=['object']).columns:
    household_df[col] = (household_df[col].where(household_df[col].notna())  
        .str.lower().str.strip().str.replace(r'\s+', ' ', regex=True)
        .str.replace(r'[^a-z0-9 /\-]', '', regex=True))
    
household_df.head()
```

```python
# Drops unnecessary column (Unnamed is an index duplication column)
household_df = household_df.drop('Unnamed: 0', axis=1)
```

```python
# Renames columns for readability
household_df.columns = ['Sex', 'Age', 'Status', 'Education', 
                        'With Job or Business', 'Occupation', 'Worker Class', 
                        'Household Type', 'Family Size', 'Less Than 5y', 
                        '5y to 17y', 'Total Employed', 'Food Cost', 
                        'Bread and Cereal','Rice', 'Meat', 'Fish', 'Fruit', 
                        'Vegetables', 'Restaurant and Hotel', 'Alcohol', 
                        'Tobacco', 'Clothing', 'Housing and Water', 
                        'Medical Care', 'Transportation','Communication', 
                        'Education Cost','Misc', 'Special Occasions']
household_df.head()
```

## Convert categorical datatypes  
Rather than dummy code my categorical variables, I will be using the enable_categorical=True argument when initializing my XGBClassifier. This argument allows the model to handle categorical variables in their text form. The XGB model can handle both category and int64 dtypes within one model, but you have to convert the object dtypes to categorical for this to work. 

```python
# Stores categorical columns 
categorical_cols = household_df.select_dtypes(include=['object']).columns

# Loops through categorical columns, assigns new 'category' dtype
for col in categorical_cols:
    household_df[col] = household_df[col].astype('category')

# Outputs new column dtypes
household_df.info()
```

## Encode Target Variable
XGBClassifier wants the target variable represented as 0 and 1 integers, so I will use mapping to change this within my dataset. I'm setting extended family as positive and single family as negative, since the goal of our model is to predict extended family households. 

```python
# Creates mapping dictionary for target encoding
mapping = {'single family': 0, 'extended family': 1}

# Replaces binary values in target variable with mapped values
household_df['Household Type'] = household_df['Household Type'].map(mapping)

# Converts dtype to a format the model will understand
household_df['Household Type'] = household_df['Household Type'].astype('int64')
household_df.info()
```

## Split training and test sets
In order to prevent data leakage, I will be splitting my data into training and test sets before conducting any feature transformation. 

```python
from sklearn.model_selection import train_test_split

# Isolate dependent and target variables 
X, y = household_df.drop('Household Type', axis=1), household_df['Household Type']
```

```python
# Randomly splits 30% of the data into the test set, and 70% into the training
# set, uses random_state for reproducibility
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, 
                                                    random_state=1)
```

## Feature Transformation
The numerical descriptive statistics show that we are dealing with very different scales between the earlier and later numerical features, as well as some outliers evident by the mean and maximum values. The XGBoost model is robust to both, but normalizing the skewed distributions may lead to a more efficient model. I decided to use the log transformation on the skewed expenditure data only (the other numerical data didn't need it). 

```python
# Outputs descriptive statistics for numerical data
household_df.describe()
```

```python
# Outputs descriptive statistics for categorical data
household_df.describe(exclude=np.number)
```

```python
# Stores list of expenditure column
expenditure_cols = ['Food Cost', 'Bread and Cereal', 'Rice', 'Meat', 'Fish',
                    'Fruit', 'Vegetables', 'Restaurant and Hotel', 'Alcohol', 
                    'Tobacco', 'Clothing', 'Housing and Water', 'Medical Care', 
                    'Transportation', 'Communication', 'Education Cost', 
                    'Misc', 'Special Occasions']

# Uses log transformation on training set
X_train[expenditure_cols] = X_train[expenditure_cols].apply(lambda x: np.log1p(x))

# Applies same transformation on test set
X_test[expenditure_cols] = X_test[expenditure_cols].apply(lambda x: np.log1p(x))
```

```python
X_train.describe()
```

```python
X_test.describe()
```

## Feature Selection 
XGBoost uses embedded feature selection methods. I want to take advantage of these, so I will be fitting the model to discover the most important features before the final fit and evaluation in the next milestone. The features selected coincide with discoveries in the intial EDA phase. There was minimal accuracy loss with almost half of the features removed. 

```python
from xgboost import XGBClassifier
from sklearn.feature_selection import SelectFromModel
from sklearn.metrics import accuracy_score

# Uses class values counts to calculate scale_pos_weight argument 
scale_pos_weight = ((y_train == 0).sum()) / ((y_train == 1).sum())

# Trains first XGBoost model on all features, uses enable categorical to use 
# categorical features without one-hot encoding, uses scale_pos_weight to 
# account for class imbalance 
xgb = XGBClassifier(enable_categorical=True, 
                      scale_pos_weight=scale_pos_weight)
model = xgb.fit(X_train, y_train)

# Tests model accuracy on all features
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy with all features: {accuracy:.4f}")
```

```python
# Creates feature importance selector
embedded_selector = SelectFromModel(estimator=model, threshold='median')

# Fits selector to training data  
embedded_selector.fit(X_train, y_train)

# Stores features that were selected
selected_features = X_train.columns[embedded_selector.get_support()]

# Creates new X_train and X_test dataframes with only selected features to 
# meet format expected by the model 
X_train_selected = pd.DataFrame(embedded_selector.transform(X_train), 
                                columns=selected_features)
X_test_selected = pd.DataFrame(embedded_selector.transform(X_test), 
                               columns=selected_features)

# Uses intersection to match the categorical variables from the original 
# split with the newly selected features 
categorical_cols = (X_train.select_dtypes(include=['category'])
    .columns.intersection(selected_features))

# Ensures categorical variables in selected features have categorical dtype, 
# to allow use of enable_categorical
for col in categorical_cols:
    X_train_selected[col] = X_train_selected[col].astype('category')
    X_test_selected[col] = X_test_selected[col].astype('category')

# Converts object columns with numeric data to numeric dtype, a format that the
# model will understand
for col in X_train_selected.select_dtypes(include=['object']).columns:
    X_train_selected[col] = pd.to_numeric(X_train_selected[col], 
                                          errors='coerce')
    X_test_selected[col] = pd.to_numeric(X_test_selected[col], 
                                         errors='coerce')

# Retrains final model on selected and properly formatted features
model_selected = XGBClassifier(enable_categorical=True,
    scale_pos_weight=scale_pos_weight)

# Fits final model
model_selected.fit(X_train_selected, y_train)

# Tests model accuracy with selected features
y_pred = model_selected.predict(X_test_selected)
accuracy = accuracy_score(y_test, y_pred)

# Outputs new accuracy score and list of selected features
print(f"Accuracy with selected features: {accuracy:.4f}")
print("Selected features:", selected_features.tolist())
```

I chose the XGBoost Classifier model during the data cleaning phase of the project, because I knew my dataset dealt with missing values, outliers, and an imbalanced target class. The XGBoost Classifier model can be robust to these issues and requires less preprocessing, saving valuable time for our business problem predictions. If I were to run the GridSearchCV for model selection at this point in the project, the results would be misleading, as I have prepared the dataset for a specific model. I fit the first model in the last milestone so that I could use the XGBoost Classifier embedded feature selector. The next step will be hyperparameter tuning. 

## Hyperparameter Tuning
I used the suggested parameters from the xgboosting website to run the GridSearchCV. Although we checked the accuracy of the model before and after feature selection, I used recall as the scoring metric for the hyperparameter tuning. My goal is to minimize false negatives, which is what the recall scoring metric addresses. In our business problem, if the goal is to offer resources for heads of households living in an extended family household, minimizing the false negatives will be more advantageous to our goal. The GridSearchCV will tell us the hyperparameters that deliver the best recall score as well as store the best model for us. 

```python
from sklearn.model_selection import GridSearchCV

# Uses xgboosting website suggested parameters
param_grid = {
    'max_depth': [3, 5, 7],
    'min_child_weight': [1, 3, 5],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'learning_rate': [0.01, 0.1, 0.3]
}
```

```python
# Performs grid search on recall score, to minimize false negatives
grid_search = GridSearchCV(estimator=model_selected, param_grid=param_grid, 
                           cv=3, n_jobs=-1, verbose=0, scoring='recall')
grid_search.fit(X_train_selected, y_train)

# Prints best parameters and best recall score
print(f"Best parameters: {grid_search.best_params_}")
print(f"Best recall score: {grid_search.best_score_}")
```

## Evaluation Metrics
I chose to evaluate the XGBoost Classifier model using accuracy, precision, recall and f1, because these are well-known metrics for classification model evaluation. Accuracy can be misleading when dealing with imbalanced classes, so it is important to include more than just one metric to deliver much-needed context for the model's performance. 

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Assigns the best model found to a variable
best_model = grid_search.best_estimator_

# Uses best model to predict labels on test set with selected features
y_pred = best_model.predict(X_test_selected)

# Calculates precision, recall and f1 scores
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"The accuracy score is: {accuracy:.2f}")
print(f"The precision score is: {precision:.2f}")
print(f"The recall score is: {recall:.2f}")
print(f"The F1 Score is: {f1:.2f}")
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, roc_auc_score

# Predicts probability of positive class (extended families) for each sample
y_pred_proba = best_model.predict_proba(X_test_selected)[:, 1]

# Calculates false positive rate, true positive rate, and thresholds
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

# Computes area under curve for legend 
roc_auc = roc_auc_score(y_test, y_pred_proba)

# Plots ROC curve
plt.plot(fpr, tpr, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')  
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate (Recall)')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```
