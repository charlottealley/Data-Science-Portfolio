## Merging Data and Storing in a Database/Visualizing Data

## Step 1: Load All Dataframes into One Database
I chose SQL Lite, because I enjoyed using it in this week's homework assignment. I chose to convert each milestone's dataframe into a csv file, which I then used to create the tables in the database. I used a LEFT join, because the goal was to keep every row from the first dataset, and add additional columns from the second and third datasets. 

```python
import sqlite3
import pandas as pd

with sqlite3.connect('gun_violence.db') as conn:
    # Loads csv files into dataframes 
    df1 = pd.read_csv('school_adjacent_shootings.csv')
    df2 = pd.read_csv('gun_waiting_periods.csv')
    df3 = pd.read_csv('news_api.csv')

    # Writes dataframes to SQLite tables
    df1.to_sql('school', conn, if_exists='replace', index=False)
    df2.to_sql('waiting', conn, if_exists='replace', index=False)
    df3.to_sql('news', conn, if_exists='replace', index=False)

    # Joins on 'State' columns
    sql = '''
    SELECT *
    FROM school s
    LEFT JOIN waiting w ON s.State = w.State
    LEFT JOIN news n ON s.State = n.State
    '''

    joined_df = pd.read_sql(sql, conn)

# Filters dataframe to drop duplicate 'State' columns
joined_df = joined_df.loc[:, ~joined_df.columns.duplicated()]

# Outputs first five rows
joined_df.head()
```

## Step 2: Recode and Rename 'Has Waiting Period' column
When creating the SQL database, some changes were made to the data. When looking at the info() output, the 'Has Waiting Period' column has been changed from True and False values to an int64 dtype with 0's and 1's as replacements. Since I know that I will be using these values in more than one visualization, I decided to impute them with more descriptive values that will deliver more context to the reader. 

```python
joined_df.info()
```

```python
# Maps integers to descriptive categorical labels for plotting
joined_df['Has Waiting Period'] = joined_df['Has Waiting Period'].map({
    0: 'No Waiting Period',
    1: 'Waiting Period Required'
})

# Rename column 
joined_df.rename(columns={'Has Waiting Period': 'Waiting Period Status'}, 
                 inplace=True)

joined_df.head()
```

## Step 3: Convert 'Incident Date' Data Type
I did this step in a previous notebook, but the column was converted back to an int64 dtype with the SQL merge.

```python
# Converts incident_date column to datetime format, uses coerce to create 
# Not a Time for invalid entries
joined_df['Incident_Date'] = pd.to_datetime(joined_df['Incident_Date'], 
                                           errors='coerce')
# Displays new data type
joined_df.info()
```

## Step 4: Impute missing values and Human Readable Dataset
As previously mentioned, due to the news api request limitations, not every state has been included in the article sentiment analysis average. Instead of creating a boolean column to communicate this fact, I decided to impute the missing values with an explanation. 

```python
# Imputes missing values with descriptive value 
joined_df['Article Sentiment Average'] = joined_df['Article Sentiment Average']\
    .fillna('Not Measured')

# Human readable dataset
joined_df.head()
```

## Step 5: Visualization 1: Waiting Period and Total Incidents 
27% of states in the US have a waiting period requirement to purchase some type of firearm. 42.2% of all school-adjacent shooting incidents since 2015 occured in a state that currently has a waiting period requirement. It is important to note that depending on when the waiting period was enacted, an incident could have occured in a state that now has a waiting period, but didn't have one in place at the time of the incident (for example, CO passed their gun waiting period law in 2023, but incidents exist before 2023). 

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Uses Pandas built-in pie plot for simplicity and readability
counts = joined_df['Waiting Period Status'].value_counts()

# Plots pie plot with class percentages
counts.plot(kind='pie', labels=counts.index.astype(str), autopct='%1.1f%%', 
            colors=sns.color_palette("Greys"), startangle=90)

plt.title('Total School-Adjacent Shootings\nby Gun Waiting Period Status', 
          fontsize=16)
plt.ylabel('') 
plt.show()
```

## Step 6: Visualization 2: Waiting Period and Average Totals Killed/Injured 
The box plots show that there is minimal difference in the averages of total number killed/injured in school-adjacent shootings between states that have a gun waiting period and states that don't have any restrictions. 

```python
import seaborn as sns

# Sets minimalistic style, uses grids for readability
sns.set(style="whitegrid")

# Creates one figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Creates both plots, uses standard deviations for error bars
sns.barplot(data=joined_df, x='Waiting Period Status', y='Total_Killed', 
            ax=axes[0], errorbar='sd', hue='Waiting Period Status', 
            legend=False, palette="Greys")
sns.barplot(data=joined_df, x='Waiting Period Status', y='Total_Injured', 
            ax=axes[1], errorbar='sd', hue='Waiting Period Status', 
            legend=False, palette="Greys")

# Sets titles and labels
axes[0].set_title('Gun Waiting Period Effect on Average Killed', fontsize=16)
axes[0].set_ylabel('Average Killed')
axes[0].set_xlabel('')

axes[1].set_title('Gun Waiting Period Effect on Average Injured', fontsize=16)
axes[1].set_ylabel('Average Injured')
axes[1].set_xlabel('')

# Retrievies current y-axis limits
y_min = min(axes[0].get_ylim()[0], axes[1].get_ylim()[0])
y_max = max(axes[0].get_ylim()[1], axes[1].get_ylim()[1])
# Sets both subplots to share same minimum and maximum y-axis limits 
axes[0].set_ylim(y_min, y_max)
axes[1].set_ylim(y_min, y_max)

plt.tight_layout()
plt.show()
```

## Step 7: Visualization 3: Sentiment Analysis by State and Gun Waiting Period Status
With the API limitations, this is only a snapshot in time and more data is needed to make any meaningful comparisons. 7 of the 13 states that were inluded in the API pull had a sentiment value score of 0, and New York had the highest sentiment value score with 1. Minnesota was the only state included in the pull that has a gun waiting period regulation in place. 

```python
# Column contains mixed dtypes, converts sentiment values to numeric
joined_df['Article Sentiment Average'] = pd.to_numeric(
    joined_df['Article Sentiment Average'], errors='coerce')

# Sorts filtered DataFrame by sentiment in descending order for cleaner plot
sorted_df = joined_df.dropna(subset=['Article Sentiment Average']).sort_values(
    'Article Sentiment Average', ascending=False)

plt.figure(figsize=(12, 7))

# Plots horizontal bar chart with states ordered by sentiment average
sns.barplot(
    data=sorted_df,
    x='Article Sentiment Average',
    y='State',
    hue='Waiting Period Status',
    palette='Greys',
    orient='h',
    order=sorted_df['State'] 
)

plt.title('Sentiment Analysis by State and Gun Waiting Period Status', 
         fontsize=18)
plt.xlabel('Sentiment Value (0=Negative, 1=Positive)')
plt.ylabel('State')
plt.legend(title='Waiting Period')
plt.xlim(right=1)

plt.tight_layout()
plt.show()
```

## Step 8: Visualization 4: School-Adjacent Shooting Incidents Over Time
The spike in school-adjacent shooting incidents the year COVID created shutdowns throughout the country (2020) is striking. The number of incidents has been steadily declining, and were close to pre-pandemic levels by the end of 2024. There is also a clear pattern of an increase during the hotter months within each year, indicating that the peak of violence often coincides with the time that many schools are on summer break. 

```python
# Converts 'Incident_Date' column back to datetime, since the join converted 
# it to an object data type 
joined_df['Incident_Date'] = pd.to_datetime(joined_df['Incident_Date'])

# Groups incidents by month, ignoring individual days for a smoother chart
joined_df['Month'] = joined_df['Incident_Date'].dt.to_period('M')  

# Groups dataframe by 'Month' column, counts number of incidents for each month
incidents_over_time = joined_df.groupby('Month').size()

# Plots time series of incident counts
plt.figure(figsize=(14, 6))
incidents_over_time.plot(color='grey')
plt.title("School-Adjacent Shooting Incidents Over Time", fontsize=20)
plt.xlabel("Year")
plt.ylabel("Number of Incidents")
plt.xticks()
plt.tight_layout()
plt.show()
```

## Step 9: Visualization 5: School-Adjacent Shooting Incidents Per State by Waiting Period Status
The states with the two highest number of incidents both have waiting periods in place. The states with the most incidents are also known for having the highest populations, with the exception of Maryland. The states with waiting periods exist within the full range of total incidents. 

```python
# Aggregates incident counts per state and waiting period status
incidents_per_state = (
    joined_df
    .groupby(['State', 'Waiting Period Status'])
    .size()
    .reset_index(name='Total Incidents')
)

# Sorts by total incidents for better visuals
sorted_df = incidents_per_state.sort_values('Total Incidents', ascending=False)

plt.figure(figsize=(14, 12))

# Defines same palette as previous visualization
palette = {'Waiting Period Required': 'grey', 'No Waiting Period': 
           'lightgrey'}

# Plots grouped barplot with hue based on Waiting Period Status
sns.barplot(
    data=sorted_df,
    x='Total Incidents',
    y='State',
    hue='Waiting Period Status',
    palette=palette,
    orient='h'
)

plt.title('Total School Adjacent Shooting Incidents Per State by Waiting Period\
 Status', fontsize=20)
plt.xlabel('Total Incidents Since 2015')
plt.ylabel('State')
plt.legend(title='Waiting Period Status')
plt.xlim(right=25000)

plt.tight_layout()
plt.show()
```

## Conclusion: 

By analyzing school-adjacent shooting incidents in relation to state-level gun waiting period laws and relevant news articles, I learned that 42.2% of incidents since 2015 occurred in states that now have some form of waiting period, while only 27% of states have such laws currently in place. However, the mean number of people killed or injured per incident, with or without outliers included, did not differ meaningfully between waiting period status classes. Sentiment analysis for the included states was predominantly negative or neutral with few exceptions. It is important to note that not every state was included in the sentiment analysis, and not every article was used, so any conclusions made will require more data. An interesting trend emerged around the COVID-19 pandemic: incidents spiked in 2020 but gradually declined thereafter, with regular seasonal increases during the summer months when school attendance drops, and the well-researched correlation between higher temperatures and increased violence exists. Additionally, states with the highest populations tended to have the most incidents, regardless of waiting period status.

There were numerous changes made to each dataset throughout the project. Imputing missing values, recoding variables, creating new variables, and conducting text and sentiment analysis are just a few examples of transformations used in the analysis. In cleaning and transforming each dataset, I made the assumption that the values on the original datasets were accurate, and relevant to the problem being examined. For example, deduplication was required in the first dataset, which relies on the assumption that each incident and school name was correctly entered into the dataset originally, and that their sources are reliable. Another assumption is that the most relevant and recent 100 articles matching the school shooting query are a representative sample for visualization purposes, while still noting that this information is incomplete due to the API limitations.

The first dataset was compiled from the Gun Violence Archive and the K-12 school data from the U.S. Department of Education’s National Center for Education Statistics. The data is verified by cross-checking geographic and incident information and excludes shootings without injuries or fatalities for reliability. The second dataset is sourced from worldpopulationreview.com, which collects official state firearm regulations and waiting period info. The site cites Gifford Law Center, Everytown Research and Fortune. The third dataset's API returns current and historic news articles published by over 150,000 worldwide sources. It sources its data through a combination of web scraping and news partnerships. There aren't any explicit legal or regulatory restrictions mentioned on the first or second source's websites. The third news API source has terms of use requirements that will create legal consequences if not followed, one being that any data from the API cannot be displayed publically. To comply, the dataset that has been merged only contains new variables created from the news API pull, and doesn't include any data that has been directly pulled from the API.

All datasets indicate that they have been collected in an ethical manner, as they cite relevant sources establishing their validity and have permission to publish the data. The sensitive nature of the topic is reflected in the omission of personal details on the school-adjacent shooting dataset. It is the nature of news outlets to both include and omit personal details of the victims of school-adjacent shootings, depending on the author and publication. Whether this is ethical depends on the manner in which it has been done, and must be considered on a case-by-case basis.

Moving forward with the ethical implications that we have identified, proper documentation of all transformations and assumptions is key in avoiding misinterpretations. Presenting all results with the limitations that have been identified by the source, as well as remaining sensitive to the topic at hand, will be necessary in honoring the overall goal of the analysis, which is gaining insight for future.
